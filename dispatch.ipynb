```python
# imports
import pandas as pd
from datetime import datetime
```

## Data
something something data

```python
# from csv file
df = pd.read_csv("log.csv")

# preproccesing
df['time'] = pd.to_datetime(df['time']) # convert to datetime
```

### Monthly Statistics

```python
monthly_group = df.groupby(pd.Grouper(key='time', freq='M')) # group by month

count = monthly_group.size().values # get dispatch count
count[0] += 783 # 06-2023 has 784 dispatches
count[1] += 50 # 07-2023 starts with 51 dispatches

# merge
monthly = monthly_group.sum().reset_index()
monthly['dispatches'] = count

# percentage stats
for col in ['dives', 'contracts', 'permits', 'raids']:
  name = col + '%'
  monthly[name] = (monthly[col] / monthly['dispatches'])*100
  monthly[name] = monthly[name].map('{:.2f}%'.format)

# no raids recorded in June 2023
monthly.iloc[0, 4] = float('NaN')
monthly.iloc[0, 9] = float('NaN')

# date format
monthly['time'] = monthly['time'].dt.strftime('%b %Y')

monthly
```
